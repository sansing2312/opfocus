```python
X=np.array([1,2,3,4,5])
y=np.array([3,4,2,5,6])
```

```python
#gradient descent 
def gradient_descent(X,y, lr=0.01, epochs=1000):
    n=len(X)
    w=0.0 
    b=0.0 

    for epoch in range(epochs):
        y_pred=w*X+b 
        error=y_pred-y 

        #compute gradients 
        dw=(2/n)*np.dot(error,X)
        db=(2/n)*np.sum(error)

        #update weights 
        w-=lr*dw 
        b-=lr*db 

        #log every 100 steps 
        if epoch%100==0:
              loss =np.mean(error**2)
              print(f"Epoch {epoch}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")
    return w,b 

w, b = gradient_descent(X, y, lr=0.01, epochs=1000)
print(f"\nFinal Model: y = {w:.2f}x + {b:.2f}")
```

```python
y_pred=w*X+b 

plt.scatter(X,y,label="Data")
plt.plot(X,y_pred,color='red',label='Prediction')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression with Gradient Descent")
plt.legend()
plt.show()
```
