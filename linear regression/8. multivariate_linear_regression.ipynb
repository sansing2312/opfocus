```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression 

X = np.array([
 [1500, 3, 10],
    [1800, 4, 5],
    [2400, 4, 20],
    [3000, 5, 8],
    [3500, 5, 12]
])

y=np.array([400, 500, 600, 650, 700])

model = LinearRegression()
model.fit(X,y)

X_new = np.array([[2500, 4, 10]])  # new house: 2500 sqft, 4 bedrooms, 10 years old
predicted_price = model.predict(X_new)
print("Predicted price ($1000s):", predicted_price[0])

# 4. View coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```
