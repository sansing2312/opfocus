```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X=np.array([1,2,3,4,5]) 
y=np.array([3,4,2,5,6]) 
```

```python
def sgd(X,y,lr=0.01,epochs=100): 
     w= 0.0 
     b=0.0 
     n=len(X) 

     for epoch in range(epochs): 
        #shuffle data each epoch 
        indices = np.arange(n) 
        np.random.shuffle(indices) 
        for i in indices:
            xi=X[i] 
            yi=y[i] 

            y_pred=w*xi+b 
            error=y_pred-yi 

            #compute gradients 
            dw=2*xi*error
            db=2*error

            #update weights 
            w-=lr*dw 
            b-=lr*db 
        
        if epoch % 10 == 0:
            y_preds = w * X + b
            loss = np.mean((y_preds - y) ** 2)
            print(f"Epoch {epoch}: Loss = {loss:.4f}, w = {w:.4f}, b = {b:.4f}")

     return w,b 


```

```python
w, b = sgd(X, y, lr=0.01, epochs=100)
print(f"\nFinal Model: y = {w:.2f}x + {b:.2f}")
```

```python
y_pred = w * X + b

plt.scatter(X, y, label="Data")
plt.plot(X, y_pred, color="red", label="SGD Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with Stochastic Gradient Descent")
plt.legend()
plt.show()
```
